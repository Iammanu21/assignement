```python
d1 = d1.fillna(0)
```

Importing PCA from sklearn

```python
from sklearn.decomposition import PCA
```

```python
pca1 = PCA(n_components=2)
principalComponents1 = pca1.fit_transform(flo_df)
```

```python
pca1.explained_variance_ratio_
```

```python
plt.scatter(principalComponents1[:,0],principalComponents1[:,1])
```

Adding labels using the "histological_type_other" column in metadata
#colours = red:adenocarcinoma | blue:neuroendocrine | green:other types

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
d21 = le.fit_transform(d2)
d21
colour = ["red","blue","green"]
```

```python
d2_c = []
for i in range(len(d21)):
     if(d21[i]==6 or d21[i]==7 or d21[i]==13 or d21[i]== 2 or d21[i]==14 or d21[i]== 8 or d21[i]== 3 ):
               d2_c.append(0)
     elif(d21[i]==9 or d21[i]== 11 or d21[i]==1 or d21[i]==10):
               d2_c.append(1)
     elif(d21[i]==12 or d21[i]==5 or d21[i]==4):
               d2_c.append(2)
     else:
               d2_c.append(3)
```

```python
l_1 = []
l_2 =[]

for i in range(len(d2_c)):
    if(d2_c[i]!=3):
        l_1.append([principalComponents1[i,0],principalComponents1[i,1]])
        l_2.append(d2_c[i])   
```

```python
l_1 = np.asanyarray(l_1)
```

PCA plot for only the numerical data 

```python
plt.scatter(l_1[:,0],l_1[:,1], c= l_2, cmap=matplotlib.colors.ListedColormap(colour))
```
